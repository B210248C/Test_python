```python
import pandas as pd

spotify = pd.read_csv("spotify.csv")
spotify
```

```python
spotify['year'] = pd.to_datetime(spotify['track_album_release_date']).dt.year
grp_spo = spotify.groupby('year')

spo_2017 = grp_spo.get_group(2017)

spo_2017
```

```python
grp_spo_2017 = spo_2017.groupby('playlist_genre')

spo_2017_list = []

for grp, info in grp_spo_2017:
    spo_2017_dict = {}
    genre_2017 = grp_spo_2017.get_group(grp)
    spo_2017_dict["Genre"] = grp
    spo_2017_dict["Average_Loudness"] = genre_2017.loudness.mean()
    spo_2017_dict["Average_Mode"] = genre_2017['mode'].mean() 
    spo_2017_dict["Average_Speechiness"] = genre_2017.speechiness.mean()
    spo_2017_dict["Average_Acousticness"] = genre_2017.acousticness.mean()
    spo_2017_dict["Average_Instrumentalness"] = genre_2017.instrumentalness.mean()
    spo_2017_dict["Average_Liveness "] = genre_2017.liveness.mean()
    spo_2017_dict["Average_Valence"] = genre_2017.valence.mean()
    spo_2017_dict["Average_Tempo"] = genre_2017.tempo.mean()
    spo_2017_dict["Average_Duration(min)"] = genre_2017.duration_ms.mean()/ 60000
    
    spo_2017_list.append(spo_2017_dict)
    
spo_2017_df = pd.DataFrame(spo_2017_list) 
spo_2017_df

# Save the DataFrame to a CSV file
spo_2017_df.to_csv("genre_averages_2017.csv", index=False)
```
